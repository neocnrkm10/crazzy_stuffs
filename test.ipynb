```python

import time
import math
import matplotlib.pyplot as plt


def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n - 1)

def combination(n, k):
    return fact(n) // (fact(n - k) * fact(k))

def isprime(p):
    if p == 0 or p == 1:
        return False
    coefs = []
    reminders = []
    yes = True
    for i in range(1, p - 1):
        coefs.append(combination(p, i))
    for c in coefs:
        reminders.append(c % p)
    for r in reminders:
        if r != 0:
            yes = False
    return yes


def is_prime(n):
    if n <= 1:
        return False
    if n <= 3:
        return True
    if n % 2 == 0 or n % 3 == 0:
        return False
    for i in range(5, int(math.sqrt(n)) + 1, 6):
        if n % i == 0 or n % (i + 2) == 0:
            return False
    return True



inputs = list(range(10, 500, 20))
time_isprime = []
time_is_prime = []


for n in inputs:

    start = time.time()
    isprime(n)
    end = time.time()
    time_isprime.append(end - start)


    start = time.time()
    is_prime(n)
    end = time.time()
    time_is_prime.append(end - start)


plt.figure(figsize=(12, 6))
plt.plot(inputs, time_isprime, label='isprime (combinations)', marker='o', color='red')
plt.plot(inputs, time_is_prime, label='is_prime (trial division)', marker='x', color='blue')
plt.title('Prime Checking Algorithms Time Comparison')
plt.xlabel('Input Number')
plt.ylabel('Time Taken (seconds)')
plt.legend()
plt.grid(True)
plt.show()
```
